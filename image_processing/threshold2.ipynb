```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread('../images/license2.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img = cv2.resize(img, (600,200))
```

```python
plt.imshow(img, cmap='gray')
```

```python
# ret , thresh = cv2.threshold(img, 140,255,cv2.THRESH_BINARY)
thresh = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 21, 2)
```

```python
plt.imshow(thresh, cmap='gray')
```

```python
kernel = np.ones((2,2), dtype=np.uint8)
kernel2 = np.ones((5,5), dtype=np.uint8)
```

```python
# removing the noise from the hello text
opening = cv2.morphologyEx(thresh,cv2.MORPH_OPEN,kernel)

plt.imshow(opening, cmap='gray')
```

```python
erode_result = cv2.dilate(thresh,kernel,iterations=5)
```

```python
plt.imshow(erode_result,cmap='gray')
```
