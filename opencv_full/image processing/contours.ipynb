```python
import cv2
import matplotlib.pyplot as plt 
import numpy as np
```

# Contours
* Contours can be explained simply as a curve joining all the continuous points along the boundary having same color or intensity.
* Contour are used for shape analysis and object detection and recognition


```python
# Draw Countours
# cv2.drawContours(image, contours, -1, (0,255,0), 3)
```

```python
black_image = np.zeros(shape = (512, 512), dtype = np.uint8)
plt.imshow(black_image, cmap = "gray")
plt.axis("off")
```

```python
black_image[100:400, 100:400] = 255
plt.imshow(black_image, cmap = "gray")
```

## Find Contours

```python
contours, hierarchy = cv2.findContours(black_image, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
```

```python
# Drawing Contours
cv2.drawContours(black_image, contours, -1, (0, 255, 0), 10) 
# -1 means draw all contours
```

```python
contours, hierarchy = cv2.findContours(black_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
contours
```

## Moments
> Calculate some features like center of mass, area of object, etc.

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
gray.shape
```

```python
contours, hierarchy = cv2.findContours(gray, 1, 2)
```

```python
cnt = contours[0]
M = cv2.moments(cnt)
print(M)
```

```python
area = cv2.contourArea(cnt)
```
