```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def load_image():
    blank_image = np.zeros((600,600))
    font = cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(blank_image,text="Hello", org=(100,300), fontFace=font,fontScale=5,color=(255,255,255),thickness=25)
    return blank_image
```

```python
def display_image(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
img = load_image()
display_image(img)
```

```python
kernel = np.ones((5,5),dtype = np.uint8)
erode_result = cv2.erode(img,kernel,iterations=2)
display_image(erode_result)
```

```python
img2 = load_image()
```

```python
white_noise = np.random.randint(low=0,high=2,size=(600,600)) # generate number between 0 to 1 for noise
display_image(white_noise)
```

```python
white_noise = white_noise * 255
```

```python
noise_image = white_noise + img2
```

```python
display_image(noise_image)
```

```python
# removing the noise from the hello text
opening = cv2.morphologyEx(noise_image,cv2.MORPH_OPEN,kernel)
display_image(opening)
```

```python
img3 = load_image()
black_noise = np.random.randint(0,2,size=(600,600))
display_image(black_noise)
```

```python
black_noise = black_noise * -255
black_noise_image = black_noise + img3 # contains -255 also
black_noise_image[black_noise_image == -255] = 0 # foreground noise will only remain now
display_image(black_noise_image)
```

```python
# removing the foreground noise
closing = cv2.morphologyEx(black_noise_image, cv2.MORPH_CLOSE,kernel)
```

```python
display_image(closing)
```
