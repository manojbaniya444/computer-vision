```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import cv2
```

```python
img = cv2.imread('./images/car2.jpg')
```

```python
type(img)
img.shape
```

```python
plt.imshow(img) # showing in bluish tint because opencv and matplotlib expects in different way
# matplotlib -> Red Gren Blue
# opencv -> Blue Green Red
# so here blue and red are swapped when using the opencv and matplotlib using combiningb
```

```python
fixed_image = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(fixed_image)
```

```python
# reading as grayscale image
img_gray = cv2.imread('./images/car2.jpg', cv2.IMREAD_GRAYSCALE)
plt.imshow(img_gray, cmap="gray")
```

```python
img_gray.shape # no longer has color channels has only rows and columns i.e intensity of pixels from dark 0 to bright 255
```

```python
fixed_image.shape # height width
```

```python
# resizing images
resized_image = cv2.resize(fixed_image,(100,200)) # width height
plt.imshow(resized_image)
```

```python
w_ratio = 0.9
h_ratio = 0.1
resized_by_ratio = cv2.resize(fixed_image, (0,0), fixed_image,w_ratio,h_ratio)
# width is 90% of the original image and the height is 10% of the original
plt.imshow(resized_by_ratio)
```

```python
# flip image
# o -1 1 values
flipped_image = cv2.flip(fixed_image,0)
plt.imshow(flipped_image)
```
