```python
# drawing on the camera
import cv2

cap = cv2.VideoCapture(0)

# variables
ltop_clicked = False
bright_clicked = False
pt1 = (0,0)
pt2 = (0,0)

# draw function
def draw(event, x, y, flags, param):
    global ltop_clicked,bright_clicked,pt1,pt2
    
    if event == cv2.EVENT_LBUTTONDOWN:
        if ltop_clicked and bright_clicked:
            pt1 = (0,0)
            pt2 = (0,0)
            ltop_clicked = False
            bright_clicked = False
            
        if ltop_clicked == False:
            pt1 = (x,y)
            ltop_clicked = True
            
        elif bright_clicked == False:
            pt2 = (x,y)
            bright_clicked = True
            

cv2.namedWindow("my_window")
cv2.setMouseCallback("my_window", draw)

while True:
    ret, frame = cap.read()
    
    # check for the button status
    if ltop_clicked:
        cv2.circle(frame,pt1,10,(0,0,255),-1)
        
    if ltop_clicked and bright_clicked:
        cv2.rectangle(frame,pt1,pt2,(0,0,255),2)

    cv2.imshow("my_window", frame)
    
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break
        

cap.release()
cv2.destroyAllWindows()
```

```python
# connect web cam
import cv2

cap = cv2.VideoCapture(0)

width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

while True:
    ret, frame = cap.read()
    gray_frame = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    cv2.imshow("frame", gray_frame)
    
    if cv2.waitKey(100) & 0xFF ==ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
# load video
import cv2
import time

cap = cv2.VideoCapture('myVideo.mp4')
fps = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

if cap.isOpened() == False:
    print("Error file not found:")
    
while cap.isOpened():
    ret,frame = cap.read()
    
    if ret:
        time.sleep(1/fps)
        cv2.imshow("frame", frame)
        
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break
    else:
        break # close after the video ends

cap.release
cv2.destroyAllWindows()
```

```python
# record from webcam
import cv2

capture = cv2.VideoCapture(0)

width = int(capture.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(capture.get(cv2.CAP_PROP_FRAME_HEIGHT))

writer = cv2.VideoWriter('myVideo.mp4',cv2.VideoWriter_fourcc(*'DIVX'), 20,(width,height))


while True:
    ret, frame = capture.read()
    writer.write(frame)
    cv2.imshow('frame', frame)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

capture.release()
writer.release()
cv2.destroyALlWindows()
```
